## Topic 1. Single layered-single output neural network

In this topic, we are going to create a simplest neural network possible that is single layered single output neural network.

In the picture below we have four input neurons, four weights and one bias, which give one output value


```python
# Lets code above neural network
# Inputs may be:
#    - pixels values of the images
#    - sensor data
#    - or outputs from the previous neurons (this is obviosly for multi-layered neural network)

# Also note that we can not modify inputs according to our needs
# We have to feed inputs data into our neural network
inputs = [1, 3, 5, 2]
```

```python
# Weights are strength of the connection between neurons
# The the main role of these value is to specify how important the input neuron to the output neuron
# Unlike inputs weights must have to adjust (i.e. training of the network) according to our actual output data
# 
weights = [0.4, 0.6, -0.5, 0.2]
```

```python
# bias is adjustment factor 
bias = 4
```

```python
# output = (inputs * weights) + bias 
output = (inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + inputs[3]*weights[3]) + bias
print(output)
```

```python
inputs = [1.0, 2.0, 3.0, 2.5]

weights1 = [0.2, 0.8, -0.5, 1.0]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]

bias1 = 2.0
bias2 = 3.0
bias3 = 0.5

output = [inputs[0]*weights1[0] + inputs[1]*weights1[1] + inputs[2]*weights1[2] + inputs[3]*weights1[3] + bias1,
          inputs[0]*weights2[0] + inputs[1]*weights2[1] + inputs[2]*weights2[2] + inputs[3]*weights2[3] + bias2,
          inputs[0]*weights3[0] + inputs[1]*weights3[1] + inputs[2]*weights3[2] + inputs[3]*weights3[3] + bias3]
print(output)
```
